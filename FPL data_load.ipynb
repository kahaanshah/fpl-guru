```python
from fpl import FPL
import aiohttp

async with aiohttp.ClientSession() as session:
    fpl = FPL(session)
    players = await fpl.get_players()
    await session.close()

```

```python
class Goalkeeper():
    
    def __init__(self, player_info):
        self.first_name = player_info.first_name
        self.price = player_info.now_cost
    
```

```python
class Defender():
    
    def __init__(self, player_info):
        self.first_name = player_info.first_name
        self.price = player_info.now_cost
```

```python
class Midfielder():
    
    def __init__(self, player_info):
        self.first_name = player_info.first_name
        self.price = player_info.now_cost
```

```python
class Forward():
    
    def __init__(self, player_info):
        self.first_name = player_info.first_name
        self.price = player_info.now_cost
```

```python
def create_positions():
    goalkeepers = dict()
    defenders = dict()
    midfielders = dict()
    forwards = dict()
    positions = [goalkeepers, defenders, midfielders, forwards]
    for player in players:
        if player.element_type == 1:
            goalkeepers[player.id] = Goalkeeper(player)
        elif player.element_type == 2:
            defenders[player.id] = Defender(player)
        elif player.element_type ==3:
            midfielders[player.id] = Midfielder(player)
        elif player.element_type ==4:
            forwards[player.id] = Forward(player)
        else:
            raise ValueError("Unexpected player element type in create positions")
    return positions
```

```python
import random

def create_team(positions):
    num_gk = 2
    num_def = 5
    num_mid = 5
    num_att = 3
    team_list = []
    while not ((sum([player.price for player in team_list]) == 0 and 
          sum([player.price for player in team_list]) > 100)):
        for _ in range(num_gk):
            team_list.append(positions[0][random.choice(list(positions[0].keys()))])
        for _ in range(num_def):
            team_list.append(positions[1][random.choice(list(positions[1].keys()))])
        for _ in range(num_mid):
            team_list.append(positions[2][random.choice(list(positions[2].keys()))])
        for _ in range(num_att):
            team_list.append(positions[3][random.choice(list(positions[3].keys()))])
        print(team_list)
    return team_list
          
    
```
